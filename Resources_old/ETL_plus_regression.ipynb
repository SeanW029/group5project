```python
from path import Path
import pandas as pd
```

```python
data = Path('Resources/kc_house_data.csv')
df = pd.read_csv(data)
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.count()
```

```python
#Checking for null values
for column in df.columns:
    print(f"Column {column} has {df[column].isnull().sum()} null values")
```

```python
df['date'] = pd.to_datetime(df['date']) #Dropping date column bcz i won't gonna day and month and creating new year column
df['year'] = df['date'].dt.year
df = df.drop(["date", "sqft_lot15"],axis=1)
df
```

```python
#Checking for duplicates in id column
df.id.duplicated().sum()
```

```python
#Dropping duplicates in 'id' columns and checking shape
new_df = df.drop_duplicates(subset='id', keep='last', inplace=False)
new_df
```

```python
new_df = new_df.rename(columns={'sqft_living': 'home_size', 'sqft_lot': 'lot_size', 'sqft_basement': 'basement_size','date': 'closing_date',
                              'yr_built': 'year', 'sqft_living15': 'homereno_2015', 'sqft_above': 'above_size',
                              'sqft_lot15': 'lotreno_2015'})
new_df                         
```

```python
new_df["price"].describe()
```

```python
new_df.to_csv('kc_house_clean.csv')
```

```python
new_df
```

```python
X = new_df.drop(['price'], axis = 1)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sn
c_matrix = X.corr()
fig = plt.figure(figsize = (9,9))
sn.heatmap(c_matrix, square=True)
plt.show()
```

```python
y = new_df["price"]
```

```python
y.head()
```

```python

X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python

X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = linear_reg.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
linear_reg.score(X, y, sample_weight=None)
```
