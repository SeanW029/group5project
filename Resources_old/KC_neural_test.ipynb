### Deliverable 1: Preprocessing the Data for a Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
from path import Path
```

```python
#  Import and read the charity_data.csv.
data = Path('Resources/kc_house_data.csv')
df = pd.read_csv(data)
df.head(20)
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#import application_df = application_df.drop(columns =['EIN', 'NAME'])
df['price'].describe()
```

```python
df['price'].median()
```

```python
df['date'] = pd.to_datetime(df['date']) #Dropping date column bcz i won't gonna day and month and creating new year column
df['year'] = df['date'].dt.year
df = df.drop(["date", "sqft_lot15"],axis=1)
df
```

```python
# X = df.drop(['date', 'price', 'zipcode', 'lat', 'long'], axis = 1)
X = df.drop(['price'], axis = 1)
```

```python
y = df["price"]
```

```python
X.dtypes
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Deliverable 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 =  180
hidden_nodes_layer2 = 256
hidden_nodes_layer3 = 150

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, kernel_initializer='normal', input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, kernel_initializer='normal', activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, kernel_initializer='normal', activation="relu"))


# Fourth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, kernel_initializer='normal', activation="relu"))


# Fifth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, kernel_initializer='normal', activation="relu"))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, kernel_initializer='normal', activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="mean_absolute_error", optimizer="adam", metrics=["mean_absolute_error"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
